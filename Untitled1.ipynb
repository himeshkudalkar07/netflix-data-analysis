```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("netflix.csv")

## Basic Information
print("Dataset shape:", df.shape)
print("\nFirst 5 rows:")
print(df.head())
print("\nDataset info:")
df.info()
print("\nDescriptive statistics:")
print(df.describe(include='all'))
# Movies vs TV Shows count
plt.figure(figsize=(8, 5))
type_counts = df['type'].value_counts()
type_counts.plot(kind='bar', color=['skyblue', 'lightcoral'])
plt.title("Movies vs TV Shows")
plt.xlabel("Type")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
# Release year distribution
plt.figure(figsize=(10, 6))
release_trend = df['release_year'].value_counts().sort_index()
release_trend.plot(kind='line', marker='o', linewidth=2)
plt.title("Content Release Trend by Year")
plt.xlabel("Year")
plt.ylabel("Count")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
# Most popular genres (top 10)
plt.figure(figsize=(12, 6))
genre_counts = df['listed_in'].value_counts().head(10)
genre_counts.plot(kind='bar', color='gold')
plt.title("Top 10 Genres on Netflix")
plt.xlabel("Genre")
plt.ylabel("Count")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
# Top countries by content count
plt.figure(figsize=(12, 6))
country_counts = df['country'].value_counts().head(10)
country_counts.plot(kind='bar', color='lightgreen')
plt.title("Top 10 Countries by Content Count")
plt.xlabel("Country")
plt.ylabel("Count")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
